```python
video= cv2.VideoCapture(0) # panggil camera

# atur size
width= int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height= int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))

writer= cv2.VideoWriter('basicvideo.mp4', cv2.VideoWriter_fourcc(*'DIVX'), 20, (width,height)) # simpan video dengan format mp4


while True:
    ret,frame= video.read() # read video

    writer.write(frame) # buat frame

    cv2.imshow('frame', frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

# simpan video
video.release()
writer.release() 
cv2.destroyAllWindows()
```

PEMODELAN

```python
# algoritma untuk face detection
face_classifier = cv2.CascadeClassifier(r'C:\Users\ACER\Downloads\kue_lapis\video\haarcascade_frontalface_default.xml')
#from keras.applications.resnet50 import ResNet50
# model = ResNet50(weights='imagenet')
# model.save('Emotion_training.h5')
classifier =load_model('Emotion_training.h5')

# labels emosi
class_labels = ['Happy','Sad']
```

PROSES PREDIKSI

```python
# Masukan semua emosi yang di deteksi ke array emotions
emotions=[] 
i=0

while True:
    try:
        print("processing "+str(i))

        file=(r"C:\Users\ACER\Downloads\kue_lapis\video\data\frame"+str(i)+".jpg") # load frame image
        img=cv2.imread(file)
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # perlu load image dalam grayscale mode
        i+=1

        faces=face_classifier.detectMultiScale(gray,1.1,4) # If wajah di temukan, it returns the positions of detected faces as Rect(x,y,w,h).
        for(x,y,w,h) in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
            # Once we get these locations, we can create a ROI for the face
            roi_gray = gray[y:y+h,x:x+w]
            roi_gray = cv2.resize(roi_gray,(48,48),interpolation=cv2.INTER_AREA)

            if np.sum([roi_gray])!=0:
                roi = roi_gray.astype('float')/255.0
                roi = img_to_array(roi)
                roi = np.expand_dims(roi,axis=0)
                preds = classifier.predict(roi)[0] # melakukan prediksi berdasarakan roi
                label=class_labels[preds.argmax()] # melkuakan pelebelan
                emotions.append(label) # tambahkan ke dalam arr emotions
            else:
                continue

        cv2.waitKey(0)
        cv2.destroyAllWindows()
    except:
        "error in image"
        break
```
